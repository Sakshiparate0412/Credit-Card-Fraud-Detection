```python
# scaling Time and Amount 
from sklearn.preprocessing import StandardScaler
credit_card_data["Amount"]= StandardScaler().fit_transform(credit_card_data["Amount"].values.reshape(-1, 1))
credit_card_data["Time"]= StandardScaler().fit_transform(credit_card_data["Time"].values.reshape(-1, 1))
```

```python
credit_card_data[["Time","Amount"]].describe()
```

```python
X=credit_card_data.drop("Class", axis=1)
y=credit_card_data["Class"].values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train, y_train)
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)
print(report)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_pred)
```

```python
 from sklearn.metrics import precision_score, recall_score
print(precision_score(y_test,y_pred))
print(recall_score(y_test,y_pred))
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,y_pred)
```

```python
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train, y_train)
y_pred2 = sgd_clf.predict(X_test)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_pred2)
```

```python
from sklearn.metrics import precision_score, recall_score
print(precision_score(y_test,y_pred2))
print(recall_score(y_test,y_pred2))
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,y_pred2)
```

```python
from sklearn.metrics import fbeta_score

f1 = fbeta_score(y_test, y_pred2, beta=1)
f2 = fbeta_score(y_test, y_pred2, beta=2)
f3 = fbeta_score(y_test, y_pred2, beta=3)
f4 = fbeta_score(y_test, y_pred2, beta=4)
f5 = fbeta_score(y_test, y_pred2, beta=5)
f6 = fbeta_score(y_test, y_pred2, beta=6)

print("F1:", f1)
print("F2:", f2)
print("F3:", f3)
print("F4:", f4)
print("F5:", f5)
print("F6:", f6)
```
